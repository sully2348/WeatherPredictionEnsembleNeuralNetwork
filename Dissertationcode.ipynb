#### The next step is to the drought index using SPI by Utilizing the precipitation column

```python
import pandas as pd
import numpy as np
from scipy.stats import gamma
```

```python
def calculate_spi(df, time_scale):
    # Calculate the monthly precipitation totals
    df['P'] = df['PRCP'].rolling(time_scale, min_periods=1).sum()

    # Calculate the long-term mean precipitation for the given time scale
    mu = df['P'].mean()

    # Calculate the long-term standard deviation of precipitation for the given time scale
    sigma = df['P'].std()

    # Calculate the SPI for the given time scale
    df['SPI'] = (df['P'] - mu) / sigma

    return df[['DATE', 'TMAX', 'TMIN', 'PRCP', 'PET', 'SPI']]
```

```python
spi_df = pd.DataFrame(columns=['DATE', 'TMAX', 'TMIN', 'PRCP', 'PET', 'SPI'])
```

```python
# Calculate the SPI for 1 month, 3 months, 6 months, and 12 months
time_scales = [1, 3, 6, 12]
for ts in time_scales:
    ts_df = calculate_spi(df, ts)
    spi_df = pd.concat([spi_df, ts_df], axis=0)

# Sort the dataframe by date
spi_df = spi_df.sort_values(by=['DATE']).reset_index(drop=True)
```

```python
def classify_drought(spi_value):
    if spi_value < -2:
        return 'Extreme drought'
    elif -2 <= spi_value < -1.5:
        return 'Severe drought'
    elif -1.5 <= spi_value < -1:
        return 'Moderate drought'
    elif -1 <= spi_value < -0.5:
        return 'Mild drought'
    elif -0.5 <= spi_value <= 0.5:
        return 'Near normal conditions'
    elif 0.5 < spi_value <= 1:
        return 'Slightly wet conditions'
    elif 1 < spi_value <= 1.5:
        return 'Moderately wet conditions'
    elif 1.5 < spi_value <= 2:
        return 'Very wet conditions'
    elif spi_value > 2:
        return 'Extremely wet conditions'

spi_df['drought index'] = spi_df['SPI'].apply(classify_drought)
```
